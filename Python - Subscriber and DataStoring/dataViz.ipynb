```python

import sqlite3
import folium
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dbName="PeopleTracking.db"
conn=sqlite3.connect(dbName)
curs=conn.cursor()
sql="""SELECT * FROM movementData"""
curs.execute(sql)
rslt=curs.fetchall()
rslt
```

```python
data=dict()
data["id"]=list()
data["latitude"]=list()
data["longitude"]=list()
data["time"]=list()
data["speed"]=list()
data["id_device"]=list()
for row in rslt:
    if row[4] > 0.0:
        data["id"].append(row[0])
        data["latitude"].append(float(row[1]))
        data["longitude"].append(float(row[2]))
        data["time"].append(row[3])
        data["speed"].append(float(row[4]))
        data["id_device"].append(row[5])

df=pd.DataFrame(data)
df
```

```python
m = folium.Map([33.32, -7.35])
for index, row in df.iterrows():
    folium.CircleMarker([row['latitude'], row['longitude']],
                        radius=2,
                        popup=row['id'],
                        fill_color="#3db7e4", # divvy color
                       ).add_to(m)
m
```

```python
# convert to (n, 2) nd-array format for heatmap
stationArr = df[['latitude', 'longitude']].as_matrix()

# plot heatmap
m.add_children(plugins.HeatMap(stationArr, radius=2))
m
```
